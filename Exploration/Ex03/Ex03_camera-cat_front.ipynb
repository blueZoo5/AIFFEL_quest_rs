```python
model_path = 'shape_predictor_68_face_landmarks.dat'
    # ì €ìž¥í•œ landmark ëª¨ë¸ì˜ ì£¼ì†Œë¥¼ model_path ë³€ìˆ˜ì— ì €ìž¥
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ shape_predictor í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ ëª¨ë¸ì„ ë¶ˆëŸ¬ì˜´
    # landmark_predictorëŠ” RGBì´ë¯¸ì§€ì™€ dlib.rectangleë¥¼ ìž…ë ¥ ë°›ê³  dlib.full_object_detectionë¥¼ ë°˜í™˜
    # dlib.rectangle: ë‚´ë¶€ë¥¼ ì˜ˆì¸¡í•˜ëŠ” ë°•ìŠ¤
    # dlib.full_object_detection: ê° êµ¬ì„± ìš”ì†Œì˜ ìœ„ì¹˜ì™€, ì´ë¯¸ì§€ ìƒì˜ ê°ì²´ì˜ ìœ„ì¹˜ë¥¼ ë‚˜íƒ€ëƒ„
print("ðŸŒ«ðŸ›¸")
```

```python
list_landmarks = []
    # ëžœë“œë§ˆí¬ì˜ ìœ„ì¹˜ë¥¼ ì €ìž¥í•  list ìƒì„±    

# ì–¼êµ´ ì˜ì—­ ë°•ìŠ¤ ë§ˆë‹¤ face landmarkë¥¼ ì°¾ì•„ëƒ…ë‹ˆë‹¤
# face landmark ì¢Œí‘œë¥¼ ì €ìž¥í•´ë‘¡ë‹ˆë‹¤
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
        # ëª¨ë“  landmarkì˜ ìœ„ì¹˜ì •ë³´ë¥¼ points ë³€ìˆ˜ì— ì €ìž¥
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        # ê°ê°ì˜ landmark ìœ„ì¹˜ì •ë³´ë¥¼ (x,y) í˜•íƒœë¡œ ë³€í™˜í•˜ì—¬ list_points ë¦¬ìŠ¤íŠ¸ë¡œ ì €ìž¥
    list_landmarks.append(list_points)
        # list_landmarksì— ëžœë“œë§ˆí¬ ë¦¬ìŠ¤íŠ¸ë¥¼ ì €ìž¥

print(len(list_landmarks[0]))
    # ì–¼êµ´ì´ nê°œì¸ ê²½ìš° list_landmarksëŠ” nê°œì˜ ì›ì†Œë¥¼ ê°–ê³ 
    # ê° ì›ì†ŒëŠ” 68ê°œì˜ ëžœë“œë§ˆí¬ ìœ„ì¹˜ê°€ ë‚˜ì—´ëœ list 
    # list_landmarksì˜ ì›ì†Œê°€ 1ê°œì´ë¯€ë¡œ list_landmarks[1]ì„ í˜¸ì¶œí•˜ë©´ IndexErrorê°€ ë°œìƒ
```

```python
# list_landmarks : 68ê°œì˜ ëžœë“œë§ˆí¬ ê°’ ì €ìž¥(ì´ëª©êµ¬ë¹„ ìœ„ì¹˜(x,y))

for dlib_rect, landmark in zip(dlib_rects, list_landmarks): # ì–¼êµ´ ì˜ì—­ì„ ì €ìž¥í•˜ê³  ìžˆëŠ” ê°’ê³¼ 68ê°œì˜ ëžœë“œë§ˆí¬ë¥¼ ì €ìž¥í•˜ê³  ìžˆëŠ” ê°’ìœ¼ë¡œ ë°˜ë³µë¬¸ ì‹¤í–‰
    print (landmark[30]) # ì½”ì˜ indexëŠ” 30 ìž…ë‹ˆë‹¤
    x = landmark[30][0] # ì´ë¯¸ì§€ì—ì„œ ì½” ë¶€ìœ„ì˜ xê°’
    y = landmark[30][1] - dlib_rect.height()//2 # ì´ë¯¸ì§€ì—ì„œ ì½” ë¶€ìœ„ì˜ yê°’ - ì–¼êµ´ ì˜ì—­ì˜ ì„¸ë¡œë¥¼ ì°¨ì§€í•˜ëŠ” í”½ì…€ì˜ ìˆ˜//2 â†’ (437, 182-(186+1//2))
    w = h = dlib_rect.width() # ì–¼êµ´ ì˜ì—­ì˜ ê°€ë¡œë¥¼ ì°¨ì§€í•˜ëŠ” í”½ì…€ì˜ ìˆ˜ (531-345+1) â†’ max(x) - min(x) +1(í”½ì…€ì˜ ìˆ˜ ì´ê¸° ë•Œë¬¸ì— 1ì„ ë”í•´ì¤Œ â†’ í”½ì…€ ìˆ˜ëŠ” ì  í•˜ë‚˜ë¡œë„ 1ì´ ë¨)
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

```python
sticker_path = '../images/cat_whiskers_transparent.png'
img_sticker = cv2.imread(sticker_path) # ìŠ¤í‹°ì»¤ ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤ // cv2.imread(ì´ë¯¸ì§€ ê²½ë¡œ) â†’ imageê°ì²´ í–‰ë ¬ì„ ë°˜í™˜
img_sticker = cv2.resize(img_sticker, (w,h)) # ìŠ¤í‹°ì»¤ ì´ë¯¸ì§€ ì¡°ì • â†’ w,hëŠ” ì–¼êµ´ ì˜ì—­ì˜ ê°€ë¡œë¥¼ ì°¨ì§€í•˜ëŠ” í”½ì…€ì˜ ìˆ˜(187) // cv2.resize(imageê°ì²´ í–‰ë ¬, (ê°€ë¡œ ê¸¸ì´, ì„¸ë¡œ ê¸¸ì´))
print (img_sticker.shape)
```

```python
# ì™¼ìª½ ëˆˆ ìœ„ì¹˜ì™€ ì˜¤ë¥¸ìª½ ëˆˆ ìœ„ì¹˜ ë³€ìˆ˜
left_eye_y = (landmark[37][1] + landmark[38][1] + landmark[39][1] + landmark[40][1]) / 4
right_eye_y = (landmark[43][1] + landmark[44][1] + landmark[45][1] + landmark[46][1]) / 4

# ëˆˆ ë†’ì´ ì°¨ì´ ê³„ì‚°
eye_diff_y = right_eye_y - left_eye_y
```

```python
# ì™¼ìª½ ëˆˆ ë†’ì´ì™€ ì˜¤ë¥¸ìª½ ëˆˆ ë†’ì´ê°€ ì–´ëŠì •ë„ ì°¨ì´ë‚˜ëŠ”ì§€ í™•ì¸ 
dx = landmark[36][0] - landmark[45][0]  # x ì°¨ì´
if dx != 0:  # x ì°¨ì´ê°€ 0ì´ë©´ íšŒì „í•  ìˆ˜ ì—†ìœ¼ë¯€ë¡œ íšŒí”¼
    angle = np.degrees(np.arctan(eye_diff_y / dx))  # íšŒì „ ê°ë„ ê³„ì‚°
else:
    angle = 0  # ê¸°ìš¸ê¸°ê°€ ì—†ìœ¼ë©´ ê°ë„ 0
```

```python
# ì´ë¯¸ì§€ ê¸°ìš¸ê¸° ì ìš© í•¨ìˆ˜
def rotate_image(image, angle):
    # ì´ë¯¸ì§€ì˜ ì¤‘ì‹¬ ê³„ì‚°
    center = (image.shape[1] // 2, image.shape[0] // 2)
    
    # íšŒì „ ë³€í™˜ í–‰ë ¬ ìƒì„± (ì¤‘ì‹¬ ê¸°ì¤€)
    rot_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    
    # ì´ë¯¸ì§€ë¥¼ íšŒì „ (ê²½ê³„ëŠ” ê·¸ëŒ€ë¡œ ìœ ì§€)
    rotated = cv2.warpAffine(image, rot_matrix, (image.shape[1], image.shape[0]), borderMode=cv2.BORDER_CONSTANT, borderValue=(0, 0, 0))
    return rotated
```

```python
# ê¸°ìš¸ì–´ì§„ ê°ë„ì— ë§žê²Œ ë³´ì •
img_sticker = rotate_image(img_sticker, angle)
```

```python
refined_x = x - w // 2 # ì½”ì—ì„œ ê·¸ë¦¼ í¬ê¸°ë§Œí¼ ì´ë™
refined_y = y  # 419
# ì›ë³¸ ì´ë¯¸ì§€ì— ìŠ¤í‹°ì»¤ ì´ë¯¸ì§€ë¥¼ ì¶”ê°€í•˜ê¸° ìœ„í•´ì„œ x, y ì¢Œí‘œë¥¼ ì¡°ì •í•©ë‹ˆë‹¤. ì´ë¯¸ì§€ ì‹œìž‘ì ì€ top-left ì¢Œí‘œì´ê¸° ë•Œë¬¸ìž…ë‹ˆë‹¤.
# ì¦‰, refined_x, refined_yê°’ì—ì„œ ì™•ê´€ ì´ë¯¸ì§€ê°€ ì‹œìž‘ë¨

print (f'(x,y) : ({refined_x},{refined_y})') 
```

```python
if refined_x < 0: 
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
# ì´ë¯¸ì§€ë¥¼ ì”Œìš°ê¸° ìœ„í•´ ì™•ê´€ ì´ë¯¸ì§€ê°€ ì‹œìž‘í•  yì¢Œí‘œ ê°’ ì¡°ì •
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :] # refined_yê°€ -98ì´ë¯€ë¡œ, img_sticker[98: , :]ê°€ ëœë‹¤. (187, 187, 3)ì—ì„œ (89, 187, 3)ì´ ë¨ (187ê°œ ì¤‘ì—ì„œ 98ê°œê°€ ìž˜ë ¤ë‚˜ê°)
    refined_y = 0

print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]

img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
print("ìŠ~")
```
